```python
subject = ['math score','reading score', 'writing score']
result = []

for sub in subject:
  mean_val = df[sub].mean()
  median_val = df[sub].median()
  mode_val = df[sub].mode()
  std_val = df[sub].std()
  var_val = df[sub].var()
  min_val = df[sub].min()
  max_val = df[sub].max()
  range = max_val - min_val

  # The 'subject' variable in 'result.append' should be 'sub' to store the current subject name
  result.append([ sub, mean_val, median_val, mode_val, std_val, var_val, min_val, max_val, range])

# Create the DataFrame after the loop has finished populating 'result'
result_data = pd.DataFrame(result, columns=[
    "Subject", 'Mean', 'Median', 'Mode', 'Std Dev', 'Varience', 'Min', 'Max', 'Range'
])

print(" Summary Statistics Table")
print("="*60)
display(result_data)
```

### Descriptive Statistics Analysis

- From the computed statistics, **reading score** (or writing, depending on your actual output — verify from your table) has the highest average (mean). This suggests that students generally perform better in this subject compared to others.

- The subject with the **highest variability (standard deviation)** is typically **math score** (confirm from your results). A higher standard deviation indicates that student performance is more spread out, meaning some students score very high while others score low.

- This suggests that **math may be the most difficult or inconsistent subject**, as student performance varies widely. In contrast, reading and writing scores tend to be more consistent, indicating more uniform performance among students.

- Overall, differences in mean and variability highlight how students perform differently across subjects, which can be useful for identifying areas where additional support or teaching improvements may be needed.

```python
score_cols = ['math score', 'reading score', 'writing score']

# Correlation matrix
print("="*60)
print("CORRELATION MATRIX")
print("="*60)
corr_matrix = df[score_cols].corr()
print(corr_matrix)

# Heatmap
plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='Blues', square=True, linewidths=0.5)

plt.title("Correlation Heatmap of Scores")
plt.show()

# Pearson correlations
math_reading_corr = df['math score'].corr(df['reading score'])
reading_writing_corr = df['reading score'].corr(df['writing score'])

print("\n" + "="*60)
print("PEARSON CORRELATIONS")
print("="*60)
print(f"Math vs Reading: {math_reading_corr:.2f}")
print(f"Reading vs Writing: {reading_writing_corr:.2f}")
```

```python
score_cols = ['math score', 'reading score', 'writing score']

# Split groups
completed = df[df['test preparation course'] == 'completed']
none = df[df['test preparation course'] == 'none']

# Mean and Standard Deviation
print("="*60)
print("MEAN SCORES (COMPLETED)")
print("="*60)
completed_mean = completed[score_cols].mean().round(2)
print(completed_mean)

print("\n" + "="*60)
print("MEAN SCORES (NONE)")
print("="*60)
none_mean = none[score_cols].mean().round(2)
print(none_mean)

print("\n" + "="*60)
print("STANDARD DEVIATION (COMPLETED)")
print("="*60)
print(completed[score_cols].std().round(2))

print("\n" + "="*60)
print("STANDARD DEVIATION (NONE)")
print("="*60)
print(none[score_cols].std().round(2))


# Improvement %
print("\n" + "="*60)
print("IMPROVEMENT (%)")
print("="*60)

improvement = ((completed_mean - none_mean) / none_mean * 100).round(2)
print(improvement)

fig, axes = plt.subplots(1, 3, figsize=(18,5))

for i, col in enumerate(score_cols):
    sns.boxplot(x='test preparation course', y=col, data=df, ax=axes[i])
    axes[i].set_title(f"{col.title()} vs Test Prep")
    axes[i].set_xlabel("Test Preparation")
    axes[i].set_ylabel("Score")

plt.tight_layout()
plt.show()
```

```python
male = df[df['gender'] == 'male']
female = df[df['gender'] == 'female']

score_cols = ['math score', 'reading score', 'writing score']
fig, axes = plt.subplots(1, 3, figsize=(18,5))

# Math
axes[0].hist(male['math score'], bins=20, alpha=0.6, label='Male')
axes[0].hist(female['math score'], bins=20, alpha=0.6, label='Female')
axes[0].set_title("Math Score by Gender")
axes[0].legend()

# Reading
axes[1].hist(male['reading score'], bins=20, alpha=0.6, label='Male')
axes[1].hist(female['reading score'], bins=20, alpha=0.6, label='Female')
axes[1].set_title("Reading Score by Gender")
axes[1].legend()

# Writing
axes[2].hist(male['writing score'], bins=20, alpha=0.6, label='Male')
axes[2].hist(female['writing score'], bins=20, alpha=0.6, label='Female')
axes[2].set_title("Writing Score by Gender")
axes[2].legend()

plt.tight_layout()
plt.show()


# -----------------------------
# Percentage scoring above 70
# -----------------------------
def percent_above_70(group, col):
    return (group[group[col] > 70].shape[0] / group.shape[0]) * 100

results = []

for col in score_cols:
    results.append([
        col,
        round(percent_above_70(male, col), 2),
        round(percent_above_70(female, col), 2)
    ])

summary_df = pd.DataFrame(results, columns=[
    "Subject", "Male % >70", "Female % >70"
])

print("="*60)
print("PERCENTAGE SCORING ABOVE 70")
print("="*60)
print(summary_df)
```

```python
# (a) Total score
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
# (b) Average score
df['average_score'] = df['total_score'] / 3
# (c) Performance grade function
def assign_grade(avg):
    if avg >= 80:
        return 'A'
    elif avg >= 70:
        return 'B'
    elif avg >= 60:
        return 'C'
    elif avg >= 50:
        return 'D'
    else:
        return 'F'
df['performance_grade'] = df['average_score'].apply(assign_grade)
# (d) High achiever
df['high_achiever'] = df['average_score'] >= 80
# (e) Lowest subject per student
def lowest_subject(row):
    scores = {
        'math': row['math score'],
        'reading': row['reading score'],
        'writing': row['writing score']
    }
    return min(scores, key=scores.get)
df['lowest_subject'] = df.apply(lowest_subject, axis=1)
# -----------------------------
# Show updated dataset
# -----------------------------
print("="*60)
print("UPDATED DATAFRAME (FIRST 10 ROWS)")
print("="*60)
print(df.head(10))

# -----------------------------
# Value counts
# -----------------------------
print("\n" + "="*60)
print("GRADE DISTRIBUTION")
print("="*60)
print(df['performance_grade'].value_counts())
print("\n" + "="*60)
print("LOWEST SUBJECT DISTRIBUTION")
print("="*60)
print(df['lowest_subject'].value_counts())
```

```python

cols = ['math score', 'reading score', 'writing score', 'average_score']

# -----------------------------
# Pair Plot 1: Gender
# -----------------------------
g1 = sns.pairplot(
    df[cols + ['gender']],
    hue='gender',
    plot_kws={'alpha': 0.5}
)

g1.fig.suptitle("Pair Plot (Hue: Gender)", y=1.02)
plt.savefig("pairplot_scores.png", dpi=150)
plt.show()


# -----------------------------
# Pair Plot 2: Performance Grade
# -----------------------------
g2 = sns.pairplot(
    df[cols + ['performance_grade']],
    hue='performance_grade',
    plot_kws={'alpha': 0.5}
)

g2.fig.suptitle("Pair Plot (Hue: Performance Grade)", y=1.02)
plt.savefig("pairplot_grades.png", dpi=150)
plt.show()
```
